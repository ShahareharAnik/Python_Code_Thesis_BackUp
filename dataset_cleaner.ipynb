```python
import os
import shutil
```

```python
new_path = r"C:\Users\hp\Desktop\sso_thesis"
folder_name = "corrected_dataset"

path = os.path.join(new_path, folder_name)
```

```python
os.mkdir(path) 
print("Directory \"{}\" created in the location \"{}\"".format(folder_name, new_path)) 
```

```python
old_path = r"C:\Users\hp\Desktop\sso_thesis\dataset csv"
os.chdir(old_path)
print(os.getcwd())
print(path)
```

```python
for yy in os.listdir():
    yy_path = os.path.join(old_path, yy)
    destination_yy = os.path.join(path, yy)
    os.mkdir(destination_yy)
    os.chdir(yy_path)
    for mm in os.listdir():
        mm_path = os.path.join(yy_path, mm)
        destination_mm = os.path.join(destination_yy, mm)
        os.mkdir(destination_mm)
        os.chdir(mm_path)
        for dd in os.listdir():
            dd_path = os.path.join(mm_path, dd)
            splitted_name = dd.split(" ")
            #print(splitted_name)
            splitted_dmy = splitted_name[-1].split("-")
            splitted_dmy[2] = splitted_dmy[2].rstrip(".csv")
            if len(splitted_dmy[2]) == 2:
                if int(splitted_dmy[2]) >= 0:
                    splitted_dmy[2] = "20" + splitted_dmy[2]
                elif int(splitted_dmy[2]) < 0:
                    splitted_dmy[2] = "19" + splitted_dmy[2]
            
            if int(splitted_dmy[0]) > 1:
                splitted_dmy[0] = str(int(splitted_dmy[0]) - 1)
            elif int(splitted_dmy[0]) == 1:
                if int(splitted_dmy[1]) == 1:
                    splitted_dmy[1] = "12"
                    splitted_dmy[2] = str(int(splitted_dmy[2]) - 1)
                else:
                    if len(splitted_dmy[1]) == 1:
                        splitted_dmy[1] = "0" + str(int(splitted_dmy[1]) - 1)
                    else:
                        splitted_dmy[1] = str(int(splitted_dmy[1]) - 1)
                
                if int(splitted_dmy[1]) == 2:
                    if (int(splitted_dmy[2])%4 == 0 and int(splitted_dmy[2])%100 != 0) or int(splitted_dmy[2])%400 == 0:
                        splitted_dmy[0] = "29"
                    else:
                        splitted_dmy[0] = "28"
                elif int(splitted_dmy[1]) == 4 or int(splitted_dmy[1]) == 6 or int(splitted_dmy[1]) == 9 or int(splitted_dmy[1]) == 11:
                    splitted_dmy[0] = "30"
                else:
                    splitted_dmy[0] = "31"
                    
            corrected_name = splitted_dmy[0] + "-" + splitted_dmy[1] + "-" + splitted_dmy[2] + ".csv"
            #print(corrected_name)
            destination_path = os.path.join(destination_mm, corrected_name)
            shutil.copy(dd_path, destination_path)
            print("Copied from \"{}\" to \"{}\"".format(dd_path, destination_path))
```

```python
print("done")
```
